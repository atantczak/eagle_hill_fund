```python
import feedparser

import requests

def fetch_medrxiv_articles():
    url = "https://api.medrxiv.org/details/medrxiv/2023-01-01/2023-01-31/0/json"
    response = requests.get(url)
    data = response.json()
    articles = data.get('collection', [])
    return articles

# Example usage
articles = fetch_medrxiv_articles()
for article in articles:
    print(f"Title: {article['title']}\nDOI: {article['doi']}\n{'-'*60}")

```
